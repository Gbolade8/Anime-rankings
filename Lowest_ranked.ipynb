```python
import pandas as pd

csv_file = '/content/animes.csv'
anime_data = pd.read_csv(csv_file)

print("First 5 rows of the dataset:")
print(anime_data.head())
```

```python
anime_data['genre'] = anime_data['genre'].str.split(', ')

exploded_genres = anime_data.explode('genre')

print("Exploded DataFrame (first 5 rows):")
print(exploded_genres.head())
```

```python
worst_rated_shows = anime_data.sort_values(by='score', ascending=True).head(30)

print("30 Worst Rated Shows:")
print(worst_rated_shows[['title', 'score']])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.barh(worst_rated_shows['title'], worst_rated_shows['score'], color='lightcoral')

plt.xlabel('Score')
plt.title('30 Worst Rated Anime Shows')

plt.gca().invert_yaxis()

plt.show()
```
